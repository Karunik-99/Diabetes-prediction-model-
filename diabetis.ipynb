```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("/content/diabetes.csv")
```

```python
data.head()
```

```python
x=data.drop(['Outcome'],axis=1)
y=data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.75)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_classifier=DecisionTreeClassifier()
```

```python
tree_classifier.fit(x_train,y_train)
```

```python
tree_predictions=tree_classifier.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(tree_predictions,y_test)
ac=accuracy_score(tree_predictions,y_test)
```

```python
print(ac)
```

```python
from sklearn.ensemble import RandomForestClassifier
forest_classifier=RandomForestClassifier()
```

```python
forest_classifier.fit(x_train,y_train)
forest_predictions=forest_classifier.predict(x_test)
```

```python
cm=confusion_matrix(forest_predictions,y_test)
ac=accuracy_score(forest_predictions,y_test)
```

```python
print(ac)
```
